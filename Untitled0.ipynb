```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

rating = pd.read_csv('https://raw.githubusercontent.com/cbikash/MREs-studies/refs/heads/main/ratings.csv')
rating.head()
```

```python
def create_matrix(df):
  user_mapper = {uid: i for i, uid in enumerate(df['userId'].unique())}
  movie_mapper = {mid: i for i, mid in enumerate(df['movieId'].unique())}
  movie_in_mapper = {i: mid for mid, i in movie_mapper.items()}
  print(user_mapper)
  print(f"movie: {movie_mapper}")
  print(f"movie mapper: {movie_in_mapper}")

  user_index = df['userId'].map(user_mapper)
  movie_index = df['movieId'].map(movie_mapper)

  X = csr_matrix((df['rating'], (movie_index, user_index)), shape=(len(movie_mapper), len(user_mapper)))

  return X, movie_in_mapper, movie_mapper

X, movie_in_mapper, movie_mapper =create_matrix(rating)

user_item_matrix = rating.pivot_table(index="title",columns="userId", values="rating")
print(user_item_matrix.iloc[:10, :5])
```

```python
def recommend_similler(movie_title, df, X, movie_in_mapper, movie_mapper, k=5):
  movie_id = df[df['title'] == movie_title]['movieId'].iloc[0]
  movie_idx = movie_mapper[movie_id]
  movie_vec = X[movie_idx]

  model = NearestNeighbors(metric='cosine', algorithm='brute')
  model.fit(X)
  distances, indices =model.kneighbors(movie_vec, n_neighbors=k+1)

  neighbors_ids = [movie_in_mapper[i] for i in indices.flatten()[1:]]
  recommendations = df[df['movieId'].isin(neighbors_ids)]['title'].unique()
  print(recommendations)

  print(f"\nBecause you liked **{movie_title}**, you might also enjoy:")
  for rec in recommendations:
      print(f"- {rec}")


recommend_similler("The Dark Knight", rating, X,
                  movie_in_mapper, movie_mapper, k=5)

```
