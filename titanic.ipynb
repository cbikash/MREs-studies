```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/cbikash/MREs-studies/main/titanic/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/cbikash/MREs-studies/main/titanic/train.csv")
data
test.head()
```

```python
data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Fare', 'Embarked'], inplace=True)
test.drop(columns=['PassengerId', 'Name', 'Ticket', 'Fare', 'Embarked', 'Cabin'], inplace=True)
```

```python
print(test.isnull().sum())
test = test.fillna({
    'Age': test['Age'].mean()
})
```

```python
data['SibSp'].value_counts()
print(data.isnull().sum())
data = data.fillna({
    "Age" : data['Age'].mean()
})
data.drop(columns=['Cabin'], inplace=True) # lots of data in cabin are null more than 50%
print(data.isnull().sum())
```

```python
from sklearn.preprocessing import  OneHotEncoder, StandardScaler, LabelEncoder

lb = LabelEncoder()
data['Sex'] = lb.fit_transform(data['Sex'])
test['Sex'] = lb.fit_transform(test['Sex'])

data

X_train = data.drop(columns=['Survived'], axis=1)
y_train = data['Survived']

X_test = test.drop(columns=['Survived'], axis=1)
y_test = test['Survived']

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

lre = LinearRegression()
lre.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100, random_state=42)
model1 = rfc.fit(X_train, y_train)
```

```python
y_pred = lre.predict(X_test)
y_pred1 = model1.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

# Calculate and print R^2 score.
r2 = r2_score(y_test, y_pred)
print(f"R-squared normal regression: {r2:.4f}")

r3 = r2_score(y_test, y_pred1)
print(f"R-squared classification: {r3:.4f}")
```
